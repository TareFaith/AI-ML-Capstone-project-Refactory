```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
```

```python
loan_type={'good': 1, 'bad':0}
def convert_good_bad(value):
    if isinstance(value, str):  # Check if value is a string
        lower_value = value.lower()  # Convert to lowercase
        return loan_type.get(lower_value, None)  # Map to numerical value
    else:
        return None  # Handle non-string values (e.g., NaN, unexpected entries)

# Apply conversion function to 'good_bad_flag' column
df['good_bad_flag'] = df['good_bad_flag'].apply(convert_good_bad)

# Print updated DataFrame
print(df)
```
